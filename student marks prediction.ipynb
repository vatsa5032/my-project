```python
import sys
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from PyQt5.QtWidgets import (
    QApplication, QWidget, QVBoxLayout, QHBoxLayout,
    QLabel, QComboBox, QPushButton, QTextEdit
)
from PyQt5.QtGui import QPalette, QLinearGradient, QColor, QBrush
from PyQt5.QtCore import QPropertyAnimation, QByteArray, QRect
from matplotlib.backends.backend_qt5agg import FigureCanvasQTAgg as FigureCanvas
from matplotlib.figure import Figure
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

class GraphCanvas(FigureCanvas):
    def __init__(self, parent=None):
        self.fig = Figure(figsize=(6, 4), dpi=100)
        self.axes = self.fig.add_subplot(111)
        super(GraphCanvas, self).__init__(self.fig)

    def plot_class_distribution(self, data, chart_type='bar'):
        self.axes.clear()
        if chart_type == 'bar':
            sns.countplot(x='Class', data=data, order=['L', 'M', 'H'], ax=self.axes, palette="Set2")
        elif chart_type == 'pie':
            counts = data['Class'].value_counts()
            self.axes.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=sns.color_palette("Set2"))
            self.axes.set_title('Class Distribution (Pie)')
        self.axes.set_ylabel("Count" if chart_type == 'bar' else "")
        self.axes.set_title("Class Distribution")
        self.draw()

class StudentPerformanceUI(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Student Performance Visualizer")
        self.setGeometry(100, 100, 850, 700)

        # Background gradient
        palette = QPalette()
        gradient = QLinearGradient(0, 0, 0, 400)
        gradient.setColorAt(0.0, QColor("#B2FEFA"))
        gradient.setColorAt(1.0, QColor("#0ED2F7"))
        palette.setBrush(QPalette.Window, QBrush(gradient))
        self.setAutoFillBackground(True)
        self.setPalette(palette)

        self.data = pd.read_csv("AI-Data.csv")
        self.chart_type = 'bar'
        self.model = self.train_model()
        self.init_ui()

    def init_ui(self):
        layout = QVBoxLayout()

        dropdown_layout = QHBoxLayout()
        self.gender_combo = QComboBox()
        self.gender_combo.addItems(["Select Gender"] + self.data["gender"].dropna().unique().tolist())
        self.grade_combo = QComboBox()
        self.grade_combo.addItems(["Select Grade"] + self.data["GradeID"].dropna().unique().tolist())
        self.absence_combo = QComboBox()
        self.absence_combo.addItems(["Select Absence"] + self.data["StudentAbsenceDays"].dropna().unique().tolist())

        dropdown_layout.addWidget(QLabel("Gender:"))
        dropdown_layout.addWidget(self.gender_combo)
        dropdown_layout.addWidget(QLabel("Grade:"))
        dropdown_layout.addWidget(self.grade_combo)
        dropdown_layout.addWidget(QLabel("Absence:"))
        dropdown_layout.addWidget(self.absence_combo)

        layout.addLayout(dropdown_layout)

        # Buttons layout
        button_layout = QHBoxLayout()

        self.button = QPushButton("Show Graph")
        self.button.clicked.connect(self.plot_graph)

        self.reset_button = QPushButton("Reset Filters")
        self.reset_button.clicked.connect(self.reset_filters)

        self.switch_button = QPushButton("Switch to Pie Chart")
        self.switch_button.clicked.connect(self.switch_chart)

        button_layout.addWidget(self.button)
        button_layout.addWidget(self.reset_button)
        button_layout.addWidget(self.switch_button)

        layout.addLayout(button_layout)

        # Graph widget
        self.graph = GraphCanvas(self)
        layout.addWidget(self.graph)

        # Prediction output
        self.prediction_label = QLabel("Prediction Output:")
        self.prediction_output = QTextEdit()
        self.prediction_output.setReadOnly(True)
        layout.addWidget(self.prediction_label)
        layout.addWidget(self.prediction_output)

        self.setLayout(layout)

    def train_model(self):
        df = self.data.copy()
        df = df.dropna()
        X = df.drop(columns=['Class'])
        y = df['Class']
        for col in X.select_dtypes(include='object'):
            le = LabelEncoder()
            X[col] = le.fit_transform(X[col])
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
        model = RandomForestClassifier()
        model.fit(X_train, y_train)
        self.feature_cols = X.columns
        self.label_encoders = {
            col: LabelEncoder().fit(self.data[col].dropna())
            for col in ['gender', 'GradeID', 'StudentAbsenceDays']
        }
        return model

    def reset_filters(self):
        self.gender_combo.setCurrentIndex(0)
        self.grade_combo.setCurrentIndex(0)
        self.absence_combo.setCurrentIndex(0)
        self.prediction_output.clear()
        self.graph.axes.clear()
        self.graph.draw()

    def switch_chart(self):
        if self.chart_type == 'bar':
            self.chart_type = 'pie'
            self.switch_button.setText("Switch to Bar Chart")
        else:
            self.chart_type = 'bar'
            self.switch_button.setText("Switch to Pie Chart")
        self.plot_graph()

    def plot_graph(self):
        gender = self.gender_combo.currentText()
        grade = self.grade_combo.currentText()
        absence = self.absence_combo.currentText()

        filtered_data = self.data.copy()
        if gender != "Select Gender":
            filtered_data = filtered_data[filtered_data["gender"] == gender]
        if grade != "Select Grade":
            filtered_data = filtered_data[filtered_data["GradeID"] == grade]
        if absence != "Select Absence":
            filtered_data = filtered_data[filtered_data["StudentAbsenceDays"] == absence]

        if filtered_data.empty:
            self.graph.axes.clear()
            self.graph.axes.text(0.5, 0.5, "No data found for selection",
                                 ha='center', va='center', fontsize=12, color='red')
            self.graph.draw()
            self.prediction_output.setText("No prediction due to empty selection.")
        else:
            # Animate
            animation = QPropertyAnimation(self.graph, b"geometry")
            animation.setDuration(400)
            animation.setStartValue(QRect(self.graph.x(), self.graph.y(), 0, 0))
            animation.setEndValue(QRect(self.graph.x(), self.graph.y(), 800, 400))
            animation.start()
            self.graph.plot_class_distribution(filtered_data, self.chart_type)
            self.show_prediction(gender, grade, absence)

    def show_prediction(self, gender, grade, absence):
        try:
            input_dict = {}
            for col in self.feature_cols:
                if col == "gender" and gender != "Select Gender":
                    input_dict[col] = [self.label_encoders['gender'].transform([gender])[0]]
                elif col == "GradeID" and grade != "Select Grade":
                    input_dict[col] = [self.label_encoders['GradeID'].transform([grade])[0]]
                elif col == "StudentAbsenceDays" and absence != "Select Absence":
                    input_dict[col] = [self.label_encoders['StudentAbsenceDays'].transform([absence])[0]]
                else:
                    input_dict[col] = [0]  # Default

            input_df = pd.DataFrame(input_dict)
            prediction = self.model.predict(input_df)[0]
            self.prediction_output.setText(f"Predicted Performance Class: **{prediction}**")
        except Exception as e:
            self.prediction_output.setText(f"Prediction error: {e}")

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = StudentPerformanceUI()
    window.show()
    sys.exit(app.exec_())
```
